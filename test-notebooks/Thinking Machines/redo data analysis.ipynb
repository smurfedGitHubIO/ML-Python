```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('./povstat_processed.csv', names=['id', 'variable', 'year', 'province', 'value', 'adm_level','region','unit'], skiprows=[0])
dataframe.head()
```

```python
del dataframe['id']
```

```python
dataframe['value'] = dataframe['value'].str.strip()
dataframe['value'] = dataframe['value'].str.replace(',', '')
dataframe['value'] = dataframe['value'].str.replace('-', '0')
dataframe['value'] = dataframe['value'].fillna(0)
dataframe['value'] = dataframe['value'].astype(np.float32)
```

```python
dataframe = dataframe.pivot(index=['variable'], columns=['year', 'province', 'region'], values='value')
```
