```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./Second Progress.xlsx', index_col='Email Address')
df.head()
```

```python
sheets_list = ['Sheet1', 'B&C', 'EdRes', 'ExteFin', 'Logistics', 'Inte']
sheets_df = {}
for i in sheets_list:
    sheets_df[i] = pd.read_excel('./Second Progress.xlsx', index_col='Email Address', sheet_name=i)
```

```python
comms = ['B&C', 'EdRes', 'ExteFin', 'Log', 'Inte']
for i in range(1, len(sheets_list)):
    # for j in sheets_df[sheets_list[i]].columns:
    sheets_df[sheets_list[i]] = df[df['Assigned Committee'] == comms[i-1]].copy(deep=True)
```
