# Preprocess Data

From the `info()` summary above, we can see that most of the data that we need doesn't contain `NULL` values. For the two columns that contains one or more `NULL` values, we can drop them as they are not necessary for any analysis that we will tackle. These two columns include `I give my consent to the UP Data Science Society to process my responses provided herewith in pursuant to the Philippine Data Privacy Act of 2012 and applicable laws.`
and `Middle Name`.

```python
df.drop(['I give my consent to the UP Data Science Society to process my responses provided herewith in pursuant to the Philippine Data Privacy Act of 2012 and applicable laws.', 'Middle Name'],axis=1,inplace=True)
```

```python
df.head()
```

We now proceed to capitalize the `Last Name`, `First Name`, and `Nickname` of each individual.

```python
names = ['Last Name', 'First Name', 'Nickname']
for name in names:
    df[name] = df[name].str.title()
df.head()
```

We proceed to check if all student numbers follow the format `XXXX-XXXXX`.

```python
df['SN Formatting'] = df['Student Number'].str.find("-") > 0
df['SN Formatting'] = df['SN Formatting'].map({True: 1, False: 0})
```

```python
df['SN Formatting']
```

```python
df['SN Formatting'].sum()
```

As we can see, all participants followed the said formatting.

We proceed to the `Personal Email` column but for this analysis, we can opt to drop this column as the UP Mail is the one that we'll be needing.

```python
df.drop('Personal Email', axis=1, inplace=True)
```

For the UP Mail, it is note-worthy that not everyone has a valid UP Mail as there are Alumnus associates. So we'll check if there are individuals which aren't Alumnus that didn't use a UP Mail for the `UP Mail` column.

```python
df[df['UP Mail'].str.find("up.edu.ph") < 0]
```
