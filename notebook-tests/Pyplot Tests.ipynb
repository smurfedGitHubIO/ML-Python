### Line plots

```python
from matplotlib import pyplot as plt
```

```python
data_x = [i for i in range(25, 36)]
data_y = [15, 27, 29, 35, 38, 44, 55, 58, 63, 68, 73]

py_data_x = [i for i in range(25, 36)]
py_data_y = [23, 34, 36, 45, 49, 56, 63, 68, 75, 80, 84]

java_data_x = [i for i in range(25, 36)]
java_data_y = [17, 30, 33, 38, 42, 48, 51, 57, 65, 72, 77]
```

```python
plt.plot(data_x, data_y, label='Average Worker')
plt.plot(py_data_x, py_data_y, label='Python')
plt.plot(java_data_x, java_data_y, label='Java')
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) per Age')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```
