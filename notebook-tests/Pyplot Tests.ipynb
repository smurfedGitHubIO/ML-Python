## Stack plots/Area plots

```python
import numpy as np
minutes = np.arange(start=1, stop=10)
player1 = [0, 1, 1, 2, 4, 5, 5, 6, 8]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]
labels=['player1', 'player2', 'player3']
```

```python
plt.stackplot(minutes, player1, player2, player3, labels=labels)
plt.legend(loc='upper left')
plt.show()
```

> Note: Stack plots shows a series representation of a bar graph

## Adding Fills

```python
import pandas as pd
df = pd.read_csv('./data/test_data_for_plotting_2.csv')
df.head()
```

```python
plt.plot(df['Age'], df['All_Devs'], color='#444444', linestyle='--', label='All Devs')
plt.plot(df['Age'], df['Python'], label='Python')

plt.fill_between(df['Age'], df['Python'], df['All_Devs'],
                 where=(df['Python']> df['All_Devs']), interpolate=True, alpha=0.25,
                label='Above Avg')
plt.fill_between(df['Age'], df['Python'], df['All_Devs'],
                 where=(df['Python']<= df['All_Devs']), interpolate=True, alpha=0.25,
                label='Below Avg')
# plt.fill_between(df['Age'], df['All_Devs'], alpha=0.25)

plt.title('Median Salary (USD) per Age')
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.show()
```
