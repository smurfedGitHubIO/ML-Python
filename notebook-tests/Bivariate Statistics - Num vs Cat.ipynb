```python
import pandas as pd

df = pd.read_csv('./data/insurance.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=df, x='charges', hue='smoker', kde=True)
plt.show()
```

```python
from scipy import stats
import statistics as stat

smoker_y = df[df['smoker'] == 'yes']['charges']
smoker_n = df[df['smoker'] == 'no']['charges']

t, p = stats.ttest_ind(smoker_y, smoker_n)
print(f't-value: {t}')
print(f'p-value: {p}')
```

```python
feature = 'smoker'
label = 'charges'

groups = df[feature].unique() # note na groups ang tawag sa kada unique category para sa isang categorical feature
grouped_values = []
for value in groups:
    grouped_values.append(df[df[feature] == value][label])
t, p = stats.ttest_ind(*grouped_values)
print(f't-value: {t}')
print(f'p-value: {p}')
```
