```python
import pandas as pd
import os
```

```python
df = pd.read_excel('./ORG MEMBERS AS OF FEB 28 2024.xlsx')
df.head()
```

```python
# FIX ORG MEMBERSHIP COLUMN
org_membership_mapper = {
    'KKM ': 'KKM'
}
df['ORG MEMBERSHIP'] = df['ORG MEMBERSHIP'].replace(org_membership_mapper)
```

```python
# FIX BARANGAY COLUMN
barangay_mapper = {
    'TAÑONG': 'TANONG',
    'TAÃ‘ONG': 'TANONG',
    'BAYAN BAYANAN': 'BAYAN-BAYANAN',
    'CONCEPCION ': 'CONCEPCION',
    'TONSUYA ': 'TONSUYA',
    'POTRERO ': 'POTRERO',
    'CATMON ': 'CATMON'
}
df['BARANGAY'] = df['BARANGAY'].replace(barangay_mapper)
```

```python
# FIX SEX COLUMN

df['SEX'].value_counts()
mapper = {
    'FEMALE': 'FEMALE',
    'F': 'FEMALE',
    'MALE': 'MALE',
    'M': 'MALE',
    'LGBT': 'LGBT',
    'Female': 'FEMALE',
    'Male': 'MALE',
    'L': 'NONE'
}

df['SEX'].fillna('L', inplace=True)
df['SEX'] = df['SEX'].map(mapper)
df['SEX'].value_counts()
```

```python
# FIX BIRTH MONTH COLUMN

df['BIRTH MONTH'].fillna('1', inplace=True)
og_lst = df['BIRTH MONTH'].value_counts().index.tolist()
lst = og_lst.copy()
df['BIRTH MONTH'].value_counts()

for i in range(len(lst)):
    if not str(lst[i]).isnumeric() or int(lst[i]) > 12:
        lst[i] = 1
    else:
        lst[i] = int(lst[i])

dct = dict(zip(og_lst, lst))

df['BIRTH MONTH'] = df['BIRTH MONTH'].map(dct)
df['BIRTH MONTH'].value_counts()
```

```python
# FIX BIRTH DAY COLUMN

df['BIRTH DAY'].fillna('1', inplace=True)
og_lst = df['BIRTH DAY'].value_counts().index.tolist()
lst = og_lst.copy()

for i in range(len(lst)):
    if not str(lst[i]).isnumeric() or int(lst[i]) > 31:
        lst[i] = 1
    else:
        lst[i] = int(lst[i])

dct = dict(zip(og_lst, lst))

df['BIRTH DAY'] = df['BIRTH DAY'].map(dct)
df['BIRTH DAY'].value_counts()
```

```python
# FIX BIRTH YEAR COLUMN
df['BIRTH YEAR'].fillna('1988', inplace=True)
qlst = ['-', '1063', '1086', '11', '1863', '1872', '1875', '1877', '1887', '1893', '1898', '197', '19778', '19788', '19825', '19888', '2993', '2022', '984', 'HAW']
test = {
    '-': '1988',
    '1063': '1963',
    '1086': '1986',
    '11': '1988',
    '1863': '1963',
    '1872': '1972',
    '1875': '1975',
    '1877': '1977',
    '1887': '1987',
    '1893': '1993', 
    '1898': '1998',
    '197': '1977',
    '19778': '1977',
    '19788': '1978',
    '19825': '1982',
    '19888': '1988',
    '2993': '1993',
    '2022': '2002',
    '984': '1984',
    'HAW': '1984'
}

df['BIRTH YEAR'] = df['BIRTH YEAR'].astype(str)
df['BIRTH YEAR'] = df['BIRTH YEAR'].replace(test)
df['BIRTH YEAR'].value_counts()
```

```python
# ADD AGE COLUMN

from datetime import datetime
df['BIRTH YEAR'] = df['BIRTH YEAR'].astype(int)
def calculate_age(row):
    dates = [31,28,31,30,31,30,31,31,30,31,30,31]
    if row['BIRTH DAY'] > row['BIRTH MONTH']:
        row['BIRTH DAY'] = row['BIRTH MONTH']
    today = datetime.today()
    birth_date = datetime(row['BIRTH YEAR'], row['BIRTH MONTH'], row['BIRTH DAY'])
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return age

df['AGE'] = df.apply(calculate_age, axis=1)

df.head()
```
