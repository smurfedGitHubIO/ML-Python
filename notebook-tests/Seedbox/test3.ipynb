```python
import pandas as pd
from PyPDF2 import PdfReader as PDFReader
import os
```

```python
dct = {}
new_lst = []
for pdfs in os.listdir('./GENDER W- TOV'):
    if pdfs[-4:] == '.pdf':
        lst = []
        cur = PDFReader(f'./GENDER W- TOV/{pdfs}')
        ctx = cur.pages[0].extract_text().split('\n')
        for i in range(12, 21):
            if i <= 15 or i == 17 or i == 19:
                lst.append(ctx[i])
            elif i == 16 or i == 18:
                lst.append(ctx[i][:-1])
                lst.append(ctx[i][-1])
            else:
                lst.append(ctx[i][:7])
                lst.append(ctx[i][7:])
        df = pd.DataFrame(lst)
        df = df.transpose()
        # print(cur.pages[0].extract_text().split('\n')[i])
        new_lst.append(df)
res_col = pd.concat(new_lst, axis=0)
```

```python
res_col.columns = ['TOTAL NUMBER OF PRECINCTS',
                  'TOTAL NUMBER OF REGISTERED MALES',
                  'TOTAL NUMBER OF REGISTERED FEMALES',
                  'TOTAL NUMBER OF REGISTERED VOTERS WITH INVALID SEX',
                  'TOTAL NUMBER OF REGISTERED VOTERS',
                   'TOTAL NUMBER OF ACTUALLY VOTED MALES',
                   'TOTAL NUMBER OF ACTUALLY VOTED FEMALES',
                   'TOTAL NUMBER OF ACTUALLY VOTED WITH INVALID SEX',
                   'TOTAL NUMBER OF VOTERS WHO ACTUALLY VOTED',
                   'PERCENTAGE OF MALE VOTED',
                   'PERCENTAGE OF FEMALE VOTED',
                   'PERCENTAGE OF ALL VOTERS',
                  ]
```
