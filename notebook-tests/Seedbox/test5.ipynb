## Mappings for age classification

```python
mapper = {
    '*B': '*',
    'C C': 'C',
    'A': 'D',
    'B': 'D',
    ' ': 'D',
    '*A': '*',
    '* *': '*',
    '*': '*',
    'C': 'C',
    'D': 'D'
}
mapper2 = {
    '*B': 'B',
    'C C': 'D',
    'A': 'A',
    'B': 'B',
    ' ': 'D',
    '*A': 'A',
    '* *': 'D',
    '*': 'D',
    'C': 'D',
    'D': 'D'
}
```

```python
dct = {}
lst = []
for csvs in os.listdir('./CSVs/'):
    df = pd.read_csv(f'./CSVs/{csvs}', encoding='latin-1')
    df = df.fillna('D')
    df['CLASSIFICATION'] = df['CLASSIFICATION'].map(mapper2)
    dct[csvs] = df
    dictionary = df['CLASSIFICATION'].value_counts().to_dict()
    if 'A' not in dictionary:
        dictionary['A'] = 0
    if 'B' not in dictionary:
        dictionary['B'] = 0
    if 'D' not in dictionary:
        dictionary['D'] = 0
        print(dictionary)
    lst.append([csvs[:-4], dictionary['A'], dictionary['B'], dictionary['D']])
```

```python
lst
```

```python
filename = './age classification.csv'
# Write the data to a CSV file
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(lst)
```
