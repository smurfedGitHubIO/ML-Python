## Mappings for age classification

```python
mapper = {
    '*B': '*',
    'C C': 'C',
    'A': 'D',
    'B': 'D',
    ' ': 'D',
    '*A': '*',
    '* *': '*',
    '*': '*',
    'C': 'C',
    'D': 'D'
}
mapper2 = {
    '*B': 'B',
    'C C': 'D',
    'A': 'A',
    'B': 'B',
    ' ': 'D',
    '*A': 'A',
    '* *': 'D',
    '*': 'D',
    'C': 'D',
    'D': 'D'
}
mapper3 = {
    '*B': '18-30 PWD',
    'C C': '60+ Unclassified',
    'A': '31-59 Illiterate',
    'B': '31-59 PWD',
    ' ': '31-59 Unclassified',
    '*A': '18-30 Illiterate',
    '* *': '18-30 Unclassified',
    '*': '18-30 Unclassified',
    'C': '60+ Unclassified',
    'D': '31-59 Unclassified'
}

mapped_values = [
    '18-30 PWD',
    '60+ Unclassified',
    '31-59 Illiterate',
    '31-59 PWD',
    '31-59 Unclassified',
    '18-30 Illiterate',
    '18-30 Unclassified',
]
```

```python
dct = {}
lst = []
district1 = "BARITAN \
BAYAN-BAYANAN \
CATMON \
CONCEPCION \
DAMPALIT \
FLORES \
HULONG DUHAT \
IBABA \
MAYSILO \
MUZON \
NIUGAN \
PANGHULO \
SAN AGUSTIN \
SANTOLAN \
TANONG \
"

district1 = district1.split()
print(district1)
for csvs in os.listdir('./CSVs/'):
    df = pd.read_csv(f'./CSVs/{csvs}', encoding='latin-1')
    df = df.fillna('D')
    df['CLASSIFICATION'] = df['CLASSIFICATION'].map(mapper)
    dct[csvs] = df
    dictionary = df['CLASSIFICATION'].value_counts().to_dict()
    # print(dictionary)
    qlst = [csvs[:-4]]
    if csvs[5:-4] not in district1:
        qlst.append(2)
    else:
        qlst.append(1)
    for vals in mapped_values:
        if vals not in dictionary:
            qlst.append(0)
        else:
            qlst.append(dictionary[vals])
    print(qlst)
    lst.append(qlst)
```

```python
filename = './test2.csv'
# Write the data to a CSV file
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(lst)
```
