```python
masterlist = pd.read_excel('./MASTERLIST 32K.xlsx')
masterlist.head()
```

```python
masterlist.drop(['NO.', 'WORK', 'CIVIL STATUS', 'BIRTH PLACE', 'PROVINCIAL ADDRESS', 'TOTAL FAMILY MEMBERS', 'EDUCATIONAL BACKGROUND', 'REMARKS', 'Unnamed: 22'], axis=1, inplace=True)
```

```python
masterlist.describe()
```

```python
masterlist['YEAR'].unique()
```

```python
invalid_years = {'1875': '1975',
                 '1887':'1987',
                 '1863': '1963',
                 '1063': '1963', 
                 '1086': '1986',
                 '984': '1984',
                 '19788': '1978',
                 '1877': '1977',
                 '19825': '1982',
                 '197': '1977',
                 '19778': '1977',
                 '19888': '1988',
                 '1898': '1998',
                 '1872': '1972',
                 '2993': '1993',
                 '-': '1988'}
masterlist['YEAR'] = masterlist['YEAR'].fillna('1988')
masterlist['YEAR'] = masterlist['YEAR'].astype('str')
masterlist['YEAR'] = masterlist['YEAR'].replace(invalid_years)
```

```python
masterlist['YEAR'] = masterlist['YEAR'].astype(int)
masterlist['AGE'] = 2024-masterlist['YEAR']
```

```python
masterlist.groupby(['ORG', 'PRECINCT #', 'BARANGAY'])['PRECINCT #'].count().to_excel('ORG-PRECINCT DISTRIBUTION.xlsx')
```

```python
masterlist.groupby(['PRECINCT #', 'AGE'])['AGE'].count().to_excel('AGE-PRECINCT DISTRIBUTION.xlsx')
```
