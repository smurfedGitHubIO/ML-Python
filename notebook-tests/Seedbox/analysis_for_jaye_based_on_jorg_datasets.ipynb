```python
list_of_jorgs = []
for ind in range(-10, -3):
    list_of_jorgs.append(comelec_malabon.columns[ind])
```

Create Turnout Percentage Column

```python
comelec_malabon['TURNOUT PERCENTAGE'] = (comelec_malabon['TURN OUT']/comelec_malabon['NO. OF REGISTERED VOTERS'])*100
```

```python
comelec_malabon['JAYE VS TO'] = (comelec_malabon['JAYE LACSON NOEL']/comelec_malabon['TURN OUT'])*100
comelec_malabon['JEANNIE VS TO'] = comelec_malabon['JEANNIE SANDOVAL']/comelec_malabon['TURN OUT']
```

```python
comelec_malabon['Influence'] = comelec_malabon['JAYE LACSON NOEL']/comelec_malabon['Total  JORG Members']
```

```python
comelec_malabon[['DISTRICT', 'BARANGAY', 'NO. OF REGISTERED VOTERS', 'TURN OUT', 'JAYE LACSON NOEL', 'CLUSTERED PRECINCT', 'TURNOUT PERCENTAGE', 'JAYE VS TO', 'LAKAS (Political)', 'Influence']].sort_values(by=['JAYE VS TO'], axis=0).to_excel('TURNOUT PERCENTAGE.xlsx')
```

```python
comelec_malabon[['DISTRICT', 'BARANGAY', 'CLUSTERED PRECINCT', 'JAYE LACSON NOEL', 'Total  JORG Members', 'Influence']].sort_values(by=['Influence'], axis=0).to_excel('MEMBERSHIP TRANSLATION.xlsx')
```

```python
comelec_malabon.columns
```

```python
barangay_vs_jorg_view = comelec_malabon.groupby(['BARANGAY'])[list_of_jorgs].sum()

for jorg in list_of_jorgs:
    comelec_malabon[f'{jorg} %'] = 0. # dummy variable

for jorg in list_of_jorgs:
    for clustered_precinct in range(0, 355):
        barangay = comelec_malabon.iloc[clustered_precinct,1]
        numerator = comelec_malabon.loc[clustered_precinct, jorg]
        denominator = barangay_vs_jorg_view.loc[barangay, jorg]
        if denominator == 0:
            comelec_malabon.at[clustered_precinct, f'{jorg} %'] = 0
        else:
            comelec_malabon.at[clustered_precinct, f'{jorg} %'] = numerator/denominator
```

```python
columns = ['KKM (Women\'s) %', 'LAKAS (Political) %', 'THUNDERS (Political) %', 
           'SANGLAHI (LGBT) %', 'BATANG LACSON (Youth) %', 'WARRIORS (Political) %']

# Find the column name with the maximum value for each row
max_column = comelec_malabon[columns].idxmax(axis=1)

comelec_malabon['MOST MEMBERS PER CLUSTERED PRECINCT'] = max_column
```

```python
comelec_malabon['MOST MEMBERS PER CLUSTERED PRECINCT'] = comelec_malabon['MOST MEMBERS PER CLUSTERED PRECINCT'].str[:-2]
```
