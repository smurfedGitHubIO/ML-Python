> All precinct info are now good to be extracted.
> Take note of the 4th index from the very first row of each dataset

```python
ans = 0
new_lst = []
dct = {} # this dictionary will contain all data per barangay
for pdfs in os.listdir('./GENDER W- TOV/'):
    if pdfs[-4:] == ".pdf":
        current_barangay = pdfs[:-4]
        dct[current_barangay] = {}
        cur = PDFReader(f'./GENDER W- TOV/{pdfs}')
        for i in range(1, len(cur.pages)):
            ctx = cur.pages[i].extract_text().split('\n')
            for j in range(17, len(ctx)-2):
                values = ctx[j].split()
                if values[0] == 'T' or values[0] == 'JETHMAL':
                    continue
                else:
                    word_counter = 0
                    for k in range(1, len(values)):
                        if not values[k].isnumeric() and values[k] != "%" and '.' not in values[k]:
                            word_counter += 1
                    while word_counter > 0:
                        word_counter -= 1
                        for ind in range(1, len(values)):
                            if not values[ind].isnumeric() and values[ind] != "%" and '.' not in values[ind]:
                                values = values[:ind] + values[ind+1:]
                                break
                    dct[current_barangay][values[0]] = values[1:]
        print(current_barangay)
        print(dct[current_barangay])
        new_df = pd.DataFrame(dct[current_barangay])
        new_df = new_df.transpose()
        dct[current_barangay] = new_df
        new_lst.append(new_df)
res_col = pd.concat(new_lst, axis=0)
```
